O Std é baseado pela média, então é calculado o quadrado da distância de cada ponto e a média e em seguida é somado, então esse valor é dividido pelo número de pontos e no final é calculado a raiz quadrada do resultado das demais operações.

Na questão da idade é representativo, pois mesmo que o max e o min sejam um pouco dispersos entre sí, a média nós mostra que naquele grupo de pessoas tem normalmente pessoas entre 28 e 34 anos, tendo uma variação de 6 anos.

6. Pois essas variáveis são qualitativas, e para realizar operações como média teria que quantificar esses dados de alguma forma.

7. A média por sí só não é um bom indicativo, pois sozinha não é possivel realizar uma analise profunda e acertiva dos dados, necessitando de outras informações, como os valores máximos e minimos, o desvio padrão e outras informações uteis. Quando maior a quantidade de informações dadas mais precisa será a análise.

```python
# Questão 8.
import statistics
medianaFilhos = dados['N de Filhos'].median()
print('mediana de Filhos\n', medianaFilhos)

medianaSal = dados['Salario (x Sal Min)'].median()
print('mediana do Salario\n', medianaSal)

medianaIdade = dados['Idade (em anos)'].median()
print('mediana da idade\n', medianaIdade)

modaFilhos = dados['N de Filhos'].mode().iloc[0]
print('\nModa de Filhos\n', modaFilhos)

modaSal = dados['Salario (x Sal Min)'].mode().iloc[0]
print('Moda de Salario\n', modaSal)

modaIdade = dados['Idade (em anos)'].mode().iloc[0]
print('Moda da idade\n', modaIdade)
```

```python
import random
import matplotlib.pyplot as plt

def roll(quantity, num_faces):
    return sum(random.randint(1, num_faces) for _ in range(quantity))

def simulate(num_simulations, quantity, num_faces):
    results = [0] * (quantity * num_faces)
    for _ in range(num_simulations):
        result = roll(quantity, num_faces)
        results[result - 1] += 1
    return results

def calculo_valor_esperado(quantity, num_faces):
    return (quantity * (num_faces + 1)) / 2

def plot_rolls(quantity, num_faces, results):
    valor_esperado = calculo_valor_esperado(quantity, num_faces)
    plt.bar(range(1, quantity * num_faces + 1), results)
    plt.axvline(valor_esperado, color='r', linestyle='--', label='Valor Esperado')
    plt.xlabel('Resultado do Lançamento')
    plt.ylabel('Frequência')
    plt.title(f'Lançamento de {quantity} dado(s) de {num_faces} faces')
    plt.legend()
    plt.show()

# Simulação para 1 rolagem de 2 faces (Moeda) 1000x
num_simulations = 1000
quantity = 1
num_faces = 2

results_1d2 = simulate(num_simulations, quantity, num_faces)
plot_rolls(quantity, num_faces, results_1d2)
```

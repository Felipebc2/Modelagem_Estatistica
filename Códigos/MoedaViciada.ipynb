```python
import random
import matplotlib.pyplot as plt

def roll(quantity, num_faces):
    return sum(random.randint(1, num_faces) for _ in range(quantity))

def simulate(num_simulations, quantity, num_faces):
    results = [0] * (quantity * num_faces)
    for _ in range(num_simulations):
        result = roll(quantity, num_faces)
        results[result - 1] += 1
    return results

def simulate_biased(num_simulations, quantity, num_faces, probabilities):
    results = [0] * (quantity * num_faces)
    for _ in range(num_simulations):
        result = roll_biased(quantity, num_faces, probabilities)
        results[result - 1] += 1
    return results

def roll_biased(quantity, num_faces, probabilities):
    result = 0
    for _ in range(quantity):
        result += random.choices(range(1, num_faces + 1), probabilities)[0]
    return result

def calculo_valor_esperado(quantity, num_faces):
    return (quantity * (num_faces + 1)) / 2

def plot_rolls(quantity, num_faces, results):
    valor_esperado = calculo_valor_esperado(quantity, num_faces)
    plt.bar(range(1, quantity * num_faces + 1), results)
    plt.axvline(valor_esperado, color='r', linestyle='--', label='Valor Esperado')
    plt.xlabel('Resultado do Lanamento')
    plt.ylabel('Frequncia')
    plt.title(f'Lanamento de {quantity} dado(s) de {num_faces} faces')
    plt.legend()
    plt.show()

# Simulao para 1 rolagem (Moeda) de 2 faces 1000x
num_simulations = 1000
quantity = 1
num_faces = 2

results_1d5 = simulate(num_simulations, quantity, num_faces)
plot_rolls(quantity, num_faces, results_1d5)

# Simulao para 1 rolagem (Moeda) de 2 faces viciada (60% - 40%) 1000x
num_simulations = 1000
quantity = 1
num_faces = 2
probabilities = [0.6, 0.4]

results_biased = simulate_biased(num_simulations, quantity, num_faces, probabilities)
plot_rolls(quantity, num_faces, results_biased)
```
